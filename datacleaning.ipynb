```python
import numpy as np
import pandas as pd
```

```python
hfi = pd.read_csv('hfi.csv')
hfi.head()
```

```python
hfi = hfi[hfi['year'] == 2016]
```

```python
hfi = hfi.rename(columns={'countries': 'country', 'pf_ss_women_missing':'missing_women',
                         'pf_ss_women':'womens_security', 'pf_movement_women':'womens_movement'})
```

```python
hfi = hfi[['country','ISO_code', 'region', 'hf_score', 'hf_rank', 'ef_score','ef_rank', 'pf_score','pf_rank', 'pf_rol_criminal',
          'missing_women','womens_security', 'womens_movement']]
hfi.head()
```

```python
gii = pd.read_csv('GII2015.csv')
gii.head()
```

```python
gii = gii.rename(columns={'HDI Rank (2015)': 'hdi_rank', 'Country':'country', '2015': '2015_gii'})
gii = gii[['country','hdi_rank', '2015_gii']]
gii.head()
# gii['2015'].isna().sum()
```

```python
femviolence = pd.read_csv('femviolence.csv')
femviolence.head()
femviolence.tail()
```

```python
# data from https://data.worldbank.org/indicator/SP.POP.TOTL
pop = pd.read_csv('population.csv')
pop.head()
```

```python
pop = pop.rename(columns={'Country Name':'country', '2021':'2021_pop'})
pop = pop[['country','2021_pop']]
pop.head()
```

```python
final = hfi.merge(gii, left_on='country', right_on='country')
final = final.merge(pop, left_on = 'country', right_on ='country')
```

```python
final.head()
```

```python
final.shape
```

```python
final_na = final.dropna()
final_na.shape
```

```python
final_na.groupby['region']
```

```python
final_na.to_csv('final_na.csv',index=False)
```
